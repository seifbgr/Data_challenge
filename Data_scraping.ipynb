```python
import requests
from bs4 import BeautifulSoup
import os

# URL du dossier GitHub avec les fichiers CSV
url_base = "https://github.com/CSSEGISandData/COVID-19/tree/master/csse_covid_19_data/csse_covid_19_daily_reports"
url_raw_base = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports/"

# Dossier local pour stocker les fichiers CSV
dossier_destination = "dossiers_des_reports_COVID"
os.makedirs(dossier_destination, exist_ok=True)

# Envoyer la requête à la page GitHub
response = requests.get(url_base)
soup = BeautifulSoup(response.text, 'html.parser')

# Trouver tous les liens de fichiers CSV dans la page
liens_csv = soup.find_all("a", href=True)
for link in liens_csv :
    if link['href'].endswith('.csv') : 
        csv_files.append(link['href'])

# Télécharger chaque fichier CSV
for csv_path in csv_files:
    # Construire l'URL brute du fichier CSV
    nom_fichier = csv_path.split("/")[-1]
    url_csv = url_raw_base + nom_fichier
    
    # Télécharger et sauvegarder le fichier CSV
    response = requests.get(url_csv)
    chemin_fichier = os.path.join(dossier_destination, nom_fichier)
    
    if response.status_code == 200:
        with open(chemin_fichier, "wb") as file:
            file.write(response.content)
        print(f"{nom_fichier} téléchargé avec succès.")
    else:
        print(f"Erreur de téléchargement pour {nom_fichier}")

print("Téléchargement terminé.")
```
